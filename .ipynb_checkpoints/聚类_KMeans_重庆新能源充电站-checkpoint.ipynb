```python
df = pd.read_csv("data/重庆充电站位置数据.csv")
```

```python
df
```

## 手肘法确定k值

```python
%matplotlib inline
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,5))
SSE = [] # 存放每次结果的误差平方和

for k in range(2,100):

    estimator = KMeans(n_clusters=k, n_init="auto") # 构造聚类器
    estimator.fit(df[["纬度","经度"]])
    SSE.append(estimator.inertia_)

X = range(2,100)

plt.xlabel('k')
plt.ylabel('SSE')
plt.xticks(X)
plt.grid(axis="both")
plt.plot(X,SSE,'o-')
plt.show()
```

## 基于K-Means的充电桩位置聚类

指定k=15，聚为15类

```python
from sklearn.cluster import KMeans
import numpy as np

model = KMeans(n_clusters=15, n_init="auto")
model.fit_transform(df[["纬度","经度"]])
model.cluster_centers_
```

### 聚类结果可视化

```python
import matplotlib.pyplot as plt
centers = model.cluster_centers_
plt.scatter(df["纬度"], df["经度"], c=model.labels_)
plt.scatter(centers[:,0], centers[:,1], marker="*", s=100, color="red")
```

## 根据轮廓系数确定K值

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
 
def test_silhouette_score(n_clusters):
    model = KMeans(n_clusters=n_clusters, n_init="auto")
    model.fit(df[["纬度","经度"]])
    y_pred = model.predict(df[["纬度","经度"]])
    # 计算轮廓分数： 所有样本点的轮廓系数的平均值
    score_mean = silhouette_score(df[["纬度","经度"]], y_pred)
    # 计算所有样本点的轮廓系数
    score_all = silhouette_samples(df[["纬度","经度"]], y_pred)
    print(f"k: {n_clusters}; 轮廓系数: {score_mean}")
 
if __name__ == '__main__':
    for i in range(2,20):
        test_silhouette_score(i)
```

聚类，聚类结果可视化略。
