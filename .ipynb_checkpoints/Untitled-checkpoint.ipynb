```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_excel(r"D:\zktr\00_Python数据可视化\源代码\health.xlsx")
df
```

```python
list(zip(df.index, df['pct_2013'], df['pct_2014']))
```

```python
def newline(p1, p2, color='black'):
    ax = plt.gca()     # 获取当前的绘图区域
    li = mlines.Line2D( [p1[0], p2[0]], [p1[1],p2[1]], color='skyblue')
    ax.add_line(li)
    return li
    
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
# 绘制散点
ax.scatter(x=df['pct_2013'], y=df.index, s=50, color='#0e668b', alpha=0.7)
ax.scatter(x=df['pct_2014'], y=df.index, s=50, color='#a3c4dc', alpha=0.7)

# 绘制线条
for y, x1, x2 in zip(df.index, df['pct_2013'], df['pct_2014']):
    newline([x1, y], [x2, y])
    
ax.set_title("2013年与2014年美国部分城市人口PCT指标的变化率", fontdict={'size':12})
ax.set_xlim(0, .25)
ax.set_xticks([.05, .1, .15, .20])
ax.set_xticklabels(['5%', '10%', '15%', '20%'])
ax.set_xlabel('变化率')
ax.set_yticks(df.index)
ax.set_yticklabels(df['city'])
ax.grid(alpha=0.5, axis='both')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
df = pd.read_excel(r'D:\zktr\00_Python数据可视化\源代码\population.xlsx')
df
```

```python
groups = df.groupby(by="Gender")
```

```python
df_male = groups.get_group("Male")
df_male
```

```python
y_tick_labels = df_male["AgeGroup"]
y_tick_labels
```

```python
[i+"岁" for i in y_tick_labels ]
```

```python
y = np.arange(1, 11)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.barh(y,df_male['Number'],label='男',color='#6699FF') 
ax.set_ylabel("年龄段（岁）")
ax.set_xticks([-100000, -75000, -50000, -25000, 0, 25000, 50000, 75000, 100000])
ax.set_xticklabels(['100000', '75000', '50000', '25000', '0', '25000', '50000', '75000', '100000'])
ax.set_xlabel("人数（个）")
ax.set_title('某城市人口金字塔')
ax.legend()
plt.show()
```
