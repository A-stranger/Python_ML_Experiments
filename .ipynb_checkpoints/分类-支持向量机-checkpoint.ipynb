## 支持向量机分类问题

$$
X = \begin{bmatrix}  
x_{11} & x_{12} & \cdots & x_{1m} \\  
x_{21} & x_{22} & \cdots & x_{2m} \\  
\vdots & \vdots & \ddots & \vdots \\  
x_{n1} & x_{n2} & \cdots & x_{nm} \\  
\end{bmatrix},  
Y = \begin{bmatrix}
y_1\\
y_2\\
\vdots \\
y_n
\end{bmatrix}
$$

$$
Y \in \{-1,1\}
$$

---

## 为什么最大间隔为1

$$
\begin{equation}  
\left\{  
\begin{aligned}  
w_1 x_1 + w_2 x_2 +b &= d, \\  
w_1 x_1 + w_2 x_2 +b &= 0, \\  
w_1 x_1 + w_2 x_2 +b &= -d. \\  
\end{aligned}  
\right. 
\end{equation}  
$$

各项均除以$d$，得：

$$
\begin{equation}  
\left\{  
\begin{aligned}  
\frac{w_1}{d} x_1 + \frac{w_2}{d} x_2  + \frac{b}{d}&= 1, \\  
\frac{w_1}{d} x_1 + \frac{w_2}{d} x_2  + \frac{b}{d}&= 1, \\  
\frac{w_1}{d} x_1 + \frac{w_2}{d} x_2  + \frac{b}{d}&= -1. \\  
\end{aligned}  
\right. 
\end{equation}  
$$

重新令 $\frac{w_1}{d} = w_1$, $\frac{w_2}{d} = w_2$, $\frac{b}{d} = b$ 得：

$$
\begin{equation}  
\left\{  
\begin{aligned}  
w_1 x_1 + w_2 x_2 +b &= 1, \\  
w_1 x_1 + w_2 x_2 +b &= 0, \\  
w_1 x_1 + w_2 x_2 +b &= -1. \\  
\end{aligned}  
\right. 
\end{equation}  
$$

---

## 支持向量到分类超平面的距离

点到直线的距离  
点：$(x_{10},x_{20})$  
直线：$w_1 x_1 + w_2 x_2 +b = 0$   
距离：

$$
d = \frac{w_1x_{10} + w_2 x_{20}+b}{\sqrt{w_1^2+w_2^2}}
$$

支持向量到分类超平面的距离

$$
d = \frac{1}{\sqrt{w_1^2+w_2^2}} = \frac{1} {{\left\lVert w \right\rVert }}
$$

通常将$\sqrt{w_1^2+w_2^2}$记为 $\left\lVert w \right\rVert$

目标为计算 $\max d$，即计算$\min\left\lVert w \right\rVert$最小时所对应的参数值$w$，

为求解方便，转为计算
$$
J(w) = \frac{1}{2} {\left\lVert w \right\rVert}^2
$$

---

## 所有训练样本点需要位于间隔两侧

$$
d = \frac{w_1x_{10} + w_2 x_{20}+b}{\sqrt{w_1^2+w_2^2}}
$$

- 正类 $w_1x_{10} + w_2 x_{20}+b \geq1$  
- 负类 $w_1x_{10} + w_2 x_{20}+b \leq-1$

$$
\begin{equation}  
\left\{  
\begin{aligned}  
w_1 x_1 + w_2 x_2 +b & \geq 1, & y_i=1\\  
w_1 x_1 + w_2 x_2 +b & \leq -1, & y_i=-1\\  
\end{aligned}  
\right. 
\end{equation}  
$$

可合并为如下不等式：

$$
y_i (w_1 x_1 + w_2 x_2 + b) >=1
$$

## 数据的标签会不会影响计算结果  

如下图

样本点$(1,3)$的标签值$y=-1$，样本点$(3,1)$的标签值$y=1$，能够被直线$x_1-x_2=0$分开，所求参数$w=[1,-1]^T$。
- 负类样本点代入直线$d<-1$；  
- 正类样本点代入直线$d>1$； 

```python
import matplotlib.pyplot as plt  
  
# 绘制直线 y = x  
x_values = [0, 5]  # x值的范围  
y_values = x_values  # 因为y = x  
plt.plot(x_values, y_values)  
plt.text(1.5, 2, "$x_1-x_2=0$")
# 绘制正样本点 (1, 3)  
plt.scatter(1, 3, marker='o', color='k', label='Negative')  
  
# 绘制负样本点 (3, 1)  
plt.scatter(3, 1, marker='+', color='r', label='Positive',s=100)  

# 添加图例  
plt.legend()  
  
# 设置x轴和y轴的标签  
plt.xlabel('$x_1$')  
plt.ylabel('$x_2$')  

plt.xlim(0,5)
plt.ylim(0,5)

# 设置x轴和y轴的比例相等，以确保直线y=x显示为对角线  
plt.gca().set_aspect('equal', adjustable='box')  
  
# 显示图形，并添加网格线  
plt.grid(True)  # 可选：添加网格线  
plt.show()
```

假如：  
样本点$(1,3)$的标签值$y=1$，样本点$(3,1)$的标签值$y=-1$，不满足直线$x_0-x_1=0$
- 负类样本点代入直线$d<-1$；  
- 正类样本点代入直线$d>1$；

此时，实际所求直线为$-x_0+x_1=0$, 参数为$w=-[1,-1]^T$。  
因此，数据标签的取值不会影响计算结果。

## 优化目标

$$
\min_w J(w) = \frac{1}{2} {\left\lVert w \right\rVert}^2. \quad s.t\quad  y_i (w^TX_i+b) >=1 \quad i=1,2,3...,N
$$

这是一个在不等式约束条件下的极值求解问题。
