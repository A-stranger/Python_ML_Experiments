## 1

$$
\begin{equation}  
\left\{  
\begin{aligned}  
w_1 x_1 + w_2 x_2 +b &= d, \\  
w_1 x_1 + w_2 x_2 +b &= 0, \\  
w_1 x_1 + w_2 x_2 +b &= -d. \\  
\end{aligned}  
\right. 
\end{equation}  
$$

$d$

$$
\begin{equation}  
\left\{  
\begin{aligned}  
\frac{w_1}{d} x_1 + \frac{w_2}{d} x_2  + \frac{b}{d}&= 1, \\  
\frac{w_1}{d} x_1 + \frac{w_2}{d} x_2  + \frac{b}{d}&= 1, \\  
\frac{w_1}{d} x_1 + \frac{w_2}{d} x_2  + \frac{b}{d}&= -1. \\  
\end{aligned}  
\right. 
\end{equation}  
$$

 $\frac{w_1}{d} = w_1$, $\frac{w_2}{d} = w_2$, $\frac{b}{d} = b$ 

$$
\begin{equation}  
\left\{  
\begin{aligned}  
w_1 x_1 + w_2 x_2 +b &= 1, \\  
w_1 x_1 + w_2 x_2 +b &= 0, \\  
w_1 x_1 + w_2 x_2 +b &= -1. \\  
\end{aligned}  
\right. 
\end{equation}  
$$

## 

  
$(x_{10},x_{20})$  
$w_1 x_1 + w_2 x_2 +b = 0$   


$$
d = \frac{w_1x_{10} + w_2 x_{20}+b}{\sqrt{w_1^2+w_2^2}}
$$



$$
d = \frac{1}{\sqrt{w_1^2+w_2^2}} = \frac{1} {{\left\lVert w \right\rVert }}
$$

$\sqrt{w_1^2+w_2^2}$ $\left\lVert w \right\rVert$

 $\max d$$\min\left\lVert w \right\rVert$$w$


$$
J(w) = \frac{1}{2} {\left\lVert w \right\rVert}^2
$$

---

## 

$$
d = \frac{w_1x_{10} + w_2 x_{20}+b}{\sqrt{w_1^2+w_2^2}}
$$

-  $w_1x_{10} + w_2 x_{20}+b \geq1$  
-  $w_1x_{10} + w_2 x_{20}+b \leq-1$

$$
\begin{equation}  
\left\{  
\begin{aligned}  
w_1 x_1 + w_2 x_2 +b & \geq 1, & y_i=1\\  
w_1 x_1 + w_2 x_2 +b & \leq -1, & y_i=-1\\  
\end{aligned}  
\right. 
\end{equation}  
$$

$$
y_i (w_1 x_1 + w_2 x_2 + b) >=1
$$
