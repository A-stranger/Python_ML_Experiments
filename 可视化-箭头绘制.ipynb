# 箭头绘制

## 箭套quiver

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.quiver(1,1,2,3, color=(1, 0, 0, 0.8), angles='xy', scale_units='xy', scale=1 )
# x = 1
# y = 1
# u = 2
# v = 3
# angles:'xy'：箭头从 (x, y) 指向 (x+u, y+v)。例如，使用它来绘制梯度场。
# To plot vectors in the x-y plane, with u and v having the same units as x and y, use angles='xy', scale_units='xy', scale=1.
# ax.quiver([0,0],[0,0],[1,0],[0,1],color=[(1, 0, 0, 0.3), (0, 1, 0, 0.3)],angles='xy', scale_units='xy', scale=1)
ax.grid()
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.quiver(0,0,1,1,color=(1, 0, 0, 0.3),angles='xy', scale_units='xy', scale=1)
ax.quiver([0,0],[0,0],[1,0],[0,1],color=[(1, 0, 0, 0.3), (0, 1, 0, 0.3)],angles='xy', scale_units='xy', scale=1)
ax.grid()
ax.set_xlabel('X')
ax.set_xlim(-1, 1)
ax.set_ylabel('Y')
ax.set_ylim(-1, 1)
plt.show()
```

## 3D Axis quiver 

```python
%matplotlib auto
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.quiver(0, 0, 0, 1, 1, 1, color=(1,0,0,0.5)) 
ax.grid()
ax.set_xlabel('X')
ax.set_xlim(-1, 1)
ax.set_ylabel('Y')
ax.set_ylim(-1, 1)
ax.set_zlabel('Z')
ax.set_zlim(-1, 1)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
x = [1,1,-1]
y = [1,-1,1]
z = [1,1,-1]

# Make a 3D quiver plot
x, y, z = np.array([[-2,0,0],[0,-2,0],[0,0,-2]])
u, v, w = np.array([[4,0,0],[0,4,0],[0,0,4]])
ax.quiver(x,y,z,u,v,w,arrow_length_ratio=0.1, color="black")
print(x,y,z,u,v,w)

ax.grid(False)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_zlim(-2,2)

plt.show()
```
